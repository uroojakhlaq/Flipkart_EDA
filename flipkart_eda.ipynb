```python
import pandas as pd
import csv
import numpy as np
```

```python
df = pd.read_csv("C:/Users/ACER/flipkart_dataset.csv")
```

```python
df.head()
```

```python
df.describe().T
```

```python
df.info()
```

```python
df.drop('image_links',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns[df.isnull().any()].tolist()
```

```python
df.dropna(subset=['selling_price'],inplace=True)
```

```python
df.dropna(subset=['mrp'],inplace=True)
```

```python
df.dropna(subset=['seller_rating'],inplace=True)
```

```python
df.shape
```

```python
df['mrp']= df['mrp'].str.replace(',','').astype(float)
```

```python
df['selling_price']= df['selling_price'].str.replace(',','').astype(float)
```

```python
df.info()
```

```python
df.assign(discount= lambda x: x.mrp-x.selling_price)
```

```python
df.info()
```

```python
df.insert(loc=7,column='discount',value=df.mrp-df.selling_price)
```

```python
df.info()
```

```python
df.head()
```

```python
df['discount_percentage']=round((df['discount']/df['mrp'])*100,2)
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
df.groupby('category_1')['category_1'].count().plot(kind='bar')
plt.title('Category with most products',fontsize=20)
plt.xlabel('Category')
plt.ylabel('Category count')
plt.show()
```

```python
#distribution of ratings
sns.histplot(df.product_rating,kde=True)
```

```python
#mean of rating
df.product_rating.mean()
```

```python
data = df.category_3.value_counts().nlargest(15)
plt.figure()
sns.barplot(x=data,y=data.index)
plt.xlabel('Count')
plt.ylabel('Categories')
plt.title('Top Selling Categories')
plt.rcParams['font.size']=15
```

```python
data = df.mrp.groupby(df.category_1).sum()
plt.figure()
plt.pie(x=data, labels=data.index,startangle=90,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title('Categories by Total revenue')
plt.tight_layout()
```

```python
data = df.product_rating.groupby(df.category_2).mean().nlargest(15)
plt.figure()
sns.barplot(x=data,y=data.index)
plt.xlabel('Rating')
plt.ylabel('category')
plt.title('Average rating for each category')
```

```python
data = df.mrp.groupby(df.title).max().nlargest(10)
plt.figure()
sns.barplot(x=data,y=data.index)
plt.xlabel('MRP')
plt.ylabel('Product')
plt.title('Product with Max Price')
```

```python
data = df.discount.groupby(df.title).max().nlargest(10)
plt.figure()
sns.barplot(x=data,y=data.index)
plt.xlabel('Discount')
plt.ylabel('Product')
plt.title('Product with Highest Discount')
```

```python
data = df.discount.groupby(df.category_1).mean()
plt.figure()
plt.pie(x=data, labels=data.index,startangle=90,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title('Categories by Average discount')
plt.tight_layout()
```
